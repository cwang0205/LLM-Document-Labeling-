# concat all extraction and export


```python
# Define the folder path
export_folder = "/content/drive/My Drive/polsky file split/"

# Get the list of all CSV files in the folder
csv_files = glob.glob(os.path.join(export_folder, "*.csv"))
# csv_files
# Initialize an empty list to store DataFrames
df_list = []

# Loop through the list of CSV files and read them
for file in csv_files:
    df = pd.read_csv(file)
    df.columns=["file_name","text","ocr_text"]
    print(df.columns)
    df_list.append(df)

# Concatenate all DataFrames into one
final_df = pd.concat(df_list, ignore_index=True)
# final_df.head()
# df_list
```

```python
export_folder="/content/drive/My Drive/polsky file split/"
output_path=os.path.join(export_folder,"final_df.csv")
final_df.to_csv(output_path,index=False)
```
